# economic PCA 후 랜덤포레스트로 kospi_after_5 예측

```python
import pandas as pd
import numpy as np
from encodings import euc_kr
import matplotlib.pyplot as plt
```

```python
ec=pd.read_csv('2019빅데이터페스티벌데이터/ECONOMIC.csv', encoding='euc_kr')
kp=pd.read_csv('2019빅데이터페스티벌데이터/KOSPI_국면분석.csv', encoding='euc_kr')
```

```python
#5일 후의 지수종가를 ex_train에 컬럼 추가
kospi_after_5 = kp['지수종가'].shift(-5)
kp['k_a_5']=kospi_after_5
```

```python
kp_1=kp[['날짜','k_a_5']]
kp_1['key']=kp.iloc[:,0].str[0:7]
kp_1.head()
```

```python
kp_2=pd.DataFrame(kp_1.groupby(['key'])['k_a_5'].mean())
```

```python
kp_2.head()
```

```python
ec['key']=ec.iloc[:,0].str[0:7]
ec.head()
```

```python
ec.iloc[:192,:]
```

```python
ec_m=pd.merge(left=kp_2, right=ec, how='left', on=['key'])
```

```python
ec_m.drop('key', axis=1)
```

```python
ec_d=ec.drop(['key','날짜'], axis=1)
kp_2=kp_2.reset_index()
```

```python
X_train=ec_d.iloc[0:191,1:]
X_test=ec_d.iloc[192:,1:]
y_train=kp_2.iloc[0:191,1]
y_test=kp_2.iloc[192:,1]
```

```python
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
#ec_m3=pd.DataFrame(ec_ms).fillna(0)
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)

X_train=np.where(np.isnan(X_train), 0, X_train)
X_test=np.where(np.isnan(X_test), 0, X_test)
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=11)

X_train=pca.fit_transform(X_train)
X_test=pca.transform(X_test)
```

```python
y_test.shape
```

```python
X_data=np.vstack([X_train,X_test])
y_target=np.hstack([y_train,y_test])
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score

# 랜덤 포레스트 학습 및 별도의 테스트 셋으로 예측 성능 평가
rf_clf = RandomForestRegressor(random_state=121)
rf_clf.fit(X_train , y_train)
pred = rf_clf.predict(X_test)

neg_mse_scores = cross_val_score(rf_clf, X_data, y_target, scoring="neg_mean_squared_error", cv = 5)
rmse_scores  = np.sqrt(-1 * neg_mse_scores)
avg_rmse = np.mean(rmse_scores)

print(' 5 교차 검증의 개별 Negative MSE scores: ', np.round(neg_mse_scores, 2))
print(' 5 교차 검증의 개별 RMSE scores : ', np.round(rmse_scores, 2))
print(' 5 교차 검증의 평균 RMSE : {0:.3f} '.format(avg_rmse))
#accuracy = accuracy_score(y_test , pred)
#print('랜덤 포레스트 정확도: {0:.4f}'.format(accuracy))
```
